# Data manipulation with Pandas

```python
import pandas as pd
```

```python
# Load the Excel file and creating a data frame
file_path = "Desktop/Main Flow/Data Cleaning and Preprocessing.xlsx"
df = pd.read_excel(file_path)
```

```python
#Print the data frame
print(df)
```

```python
# 1. Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())
```

```python
# 2. Display the column names and data types
print("\nColumn names and data types:")
print(df.dtypes)
```

```python
# 3. Filtering data based on conditions
# Example: Filter rows where Y-Kappa is greater than a value (e.g: 25)
filtered_df = df[df['Y-Kappa'] > 25]
print("\nFiltered DataFrame where Y-Kappa > 25:")
print(filtered_df)
```

```python
# 4. Handling missing values
# Example: Fill missing values with the mean of the column
df_filled = df.fillna(df.mean(numeric_only=True))
print("\nDataFrame after filling missing values with column mean:")
print(df_filled)
```

```python
# 5. Dropping rows with any missing values
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with any missing values:")
print(df_dropped)
```

```python
# 6. Calculating summary statistics
summary_stats = df.describe()
print("\nSummary statistics of the DataFrame:")
print(summary_stats)
```

```python
# 7. Selecting specific columns
selected_columns = df[['Y-Kappa', 'ChipRate']]
print("\nSelected columns (Y-Kappa and ChipRate):")
print(selected_columns)
```

```python
# 8. Sorting the DataFrame by a specific column
sorted_df = df.sort_values(by='Y-Kappa', ascending=False)
print("\nDataFrame sorted by Y-Kappa in descending order:")
print(sorted_df)
```

```python
# 9. Grouping data and calculating aggregate statistics
numeric_columns = df.select_dtypes(include='number').columns
grouped_df = df.groupby('WeakLiquorF')[numeric_columns].mean()
print("Grouped DataFrame by 'WeakLiquorF' with mean of each group:\n")
print(grouped_df)
```

```python
# 10. Creating a new column based on existing columns
df['NewColumn'] = df['Y-Kappa'] * df['ChipRate']
print("\nDataFrame with a new column 'NewColumn' (Y-Kappa * ChipRate):")
print(df.head())
```
