```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file into a Pandas DataFrame
file_path = "Desktop/Main Flow/Household_Task3.xlsx"
df = pd.read_excel(file_path)
```

```python
print(df)
```

```python
# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())
```

```python
# 1. Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(df['age'], df['income'], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Bar Chart of Age vs Income')
plt.legend(['Income'])
plt.show()
```

```python
# 2. Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df['age'], df['income'], marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Line Chart of Age vs Income')
plt.legend(['Income'])
plt.show()
```

```python
# 3. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='income', data=df, hue='expenditure', palette='viridis')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs Income')
plt.legend(title='Expenditure')
plt.show()
```

```python
# 4. Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['income'], bins=20, color='purple', alpha=0.7)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Histogram of Income')
plt.show()
```

```python
# 5. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='expenditure', y='income', data=df, palette='Set3')
plt.xlabel('Expenditure')
plt.ylabel('Income')
plt.title('Box Plot of Income by Expenditure')
plt.show()
```

```python
# 6. Pair Plot
plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='expenditure', palette='coolwarm')
plt.suptitle('Pair Plot of the DataFrame', y=1.02)
plt.show()
```
